```python
import pandas as pd

# Read the CSV file, using the first row as header and specifying the delimiter
df = pd.read_csv('/content/framingham.csv', delimiter=';')

# Now you can work with the individual columns
# Example: Display the first 5 rows of the new dataframe
df.head()

# Convert the 'age' column to integers
df['age'] = df['age'].astype(int)
df
```

```python
# prompt: plot how to find correlation betwean male ,age and Education

import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df[['male', 'age', 'education']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Male, Age, and Education')
plt.show()
```

```python
# prompt: plot calculate  average of cigsPerDay

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame and it has a 'cigsPerDay' column
# Calculate the average of 'cigsPerDay'
average_cigsPerDay = df['cigsPerDay'].mean()

# Print the average
print(f"The average number of cigarettes smoked per day is: {average_cigsPerDay}")

# Create a plot showing the distribution of 'cigsPerDay'
plt.figure(figsize=(8, 6))
sns.histplot(df['cigsPerDay'], kde=True)  # Use histplot for better visualization
plt.title('Distribution of Cigarettes Smoked Per Day')
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Frequency')
plt.axvline(average_cigsPerDay, color='red', linestyle='dashed', linewidth=2, label=f'Average: {average_cigsPerDay:.2f}')
plt.legend()
plt.show()
```

```python
# prompt: plot calculate max and min cigsPerDay

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file, using the first row as header and specifying the delimiter
df = pd.read_csv('/content/framingham.csv', delimiter=';')

# Calculate the max and min of 'cigsPerDay'
max_cigsPerDay = df['cigsPerDay'].max()
min_cigsPerDay = df['cigsPerDay'].min()

# Print the max and min
print(f"The maximum number of cigarettes smoked per day is: {max_cigsPerDay}")
print(f"The minimum number of cigarettes smoked per day is: {min_cigsPerDay}")

# Create a plot showing the distribution of 'cigsPerDay' with max and min highlighted
plt.figure(figsize=(8, 6))
sns.histplot(df['cigsPerDay'], kde=True)
plt.title('Distribution of Cigarettes Smoked Per Day')
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Frequency')
plt.axvline(max_cigsPerDay, color='red', linestyle='dashed', linewidth=2, label=f'Max: {max_cigsPerDay}')
plt.axvline(min_cigsPerDay, color='green', linestyle='dashed', linewidth=2, label=f'Min: {min_cigsPerDay}')
plt.legend()
plt.show()
```

```python
# prompt: plot column chart how to find relation betwean cigsPerDay and age

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file, using the first row as header and specifying the delimiter
df = pd.read_csv('/content/framingham.csv', delimiter=';')

# Create the column chart
plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='cigsPerDay', data=df)
plt.title('Relationship between Age and Cigarettes per Day')
plt.xlabel('Age')
plt.ylabel('Cigarettes per Day')
plt.show()
```

```python
# prompt: plot  bie  chart relation bewean  currentSmoker and  Prevalent Hyp

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file, using the first row as header and specifying the delimiter
df = pd.read_csv('/content/framingham.csv', delimiter=';')

# Create the pie chart
plt.figure(figsize=(8, 8))
df.groupby('currentSmoker')['prevalentHyp'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Relationship between Current Smoker and Prevalent Hypertension')
plt.ylabel('')  # Remove the y-axis label
plt.show()
```
